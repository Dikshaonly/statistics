```python
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt

#Random Variables and Distributions
sample = [random.randint(0,10) for _ in range(30)]
print(f"sample:{sample}")
print(f"Mean = {np.mean(sample)}")
print(f"Variance = {np.var(sample)}")

#Plotting histogram
plt.hist(sample)
plt.show()
```
